
# ___init__() - **constructor**_

The __init__() method is called automatically whenever an object is created. It initiates the object's attributes.

```python
class Student:
    def __init__ (self, name, age):
        self.name = name
        self.age = age
s1 = Student("Matthew",30)
print(s1.name)
print(s1.age)
```


# _**__str__() - String Representation**_

The __str__() method returns a user-friendly string representation of an object. It is automatically called when print() is used.

```python
class Student:
    def __init__ (self, name):
        self.name = name
    def __str__ (self):
        return f"Student {self.name}"

s = Student("Matthew")
print(s)
```


# _The __repr__() - Official Representation._

The __repr__() method returns the official or developer-friendly representation of an object. It is mainly used for debugging.


```python
class Student:
    def __init__ (self, name):
        self.name = name
    def __repr__ (self):
        return f"Student ('{self.name}')"

s  = Student("Daniella")
print(repr(s))
```


# **___len__() - length_**

The __len__() method tells Python what the length of an object is. It is called whenever len() is used.

```python
class Playlist:
    def __init__ (self,songs):
        self.songs = songs
    def __len__ (self):
        return len(self.songs)
p = Playlist(["Song1","Song2","Song3"])
print(len(p))
```


# **___eq__() - Equality Comparison_**

The __eq__() method defines how two objects are compared using the == operator.

```python
class Student:
    def __init__ (self,student_id):
        self.student_id = student_id
    def __eq__ (self,other):
        return self.student_id == other.student_id
s1 = Student(101)
s2 = Student(101)

```


# ___add__() - Addition operator_

The __add__() method defines what happens when the + operator is used on two objects.

```python
class Vector:
    def __init__ (self,x,y):
        self.x = x
        self.y = y
    def __add__ (self,other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )
    def __str__ (self):
        return f"({self.x},{self.y})"
v1 = Vector(2,3)
v2 = Vector(4,5)
print(v1 + v2)
```


# ___sub__() - Subtraction Operator_

The __sub__() method defines how objects behave when the operator is used

```python
class Vector:
    def __init__ (self,x,):
        self.x = x
    def __sub__ (self,other):
        return Vector(self.x - other.x)
    def __str__ (self):
        return str(self.x)
v1 = Vector(10)
v2 = Vector(3)
print(v1 + v2)
```


# ___mul__() - Multiplication Operator_

The __mul__() method defines how objects behave when multiplied using the * operator.

```python
class Number:
    def __init__ (self,value):
        self.value = value
    def __mul__ (self,other):
        return Number(self.value * other.value)
    def __str__ (self):
        return str(self.value)
n1 = Number(5)
n2 = Number(4)
print(n1 * n2)
```


# ___getitem__() - Item Access_

The __getitem__() method allows an object to use indexing([])like a list or tuple.

```python
class Grades:
    def __init__ (self):
        self.data = [85,90,78]
    def __getitem__ (self,index):
        return self.data[index]
g  = Grades()
print(g[0])
print(g[1])
```


# **___setitem__() - Item Assignment_**

The __setitem__() method allows values to be assigned using indexing([])

```python
class Grades:
    def __init__ (self):
        self.data = [85,90,78]
    def __setitem__ (self,index,value):
        return self.data[index]
g  = Grades()
g[1] = 100

```


# **___contains__() - Membership test_**

The __contains__() method defines how the "in" operator works for an object

```python
class Course:
    def __init__ (self):
        self.students = ["Ella","Sandra","Joe"]
    def __contains__ (self,student):
        return student in self.students
c = Course()
print("Joe" in c)
print("Gifty" in c)
```


# **___call__() - Callable Object_**

The __call__() method allows an object to behave like a function

```python
class Multiplier:
    def __init__ (self,factor):
        self.factor = factor
    def __call__ (self,number):
        return number * self.factor
double = Multiplier(2)
print(double(10))

```


# **___iter__() and __next__() - Iterator_**

These methods allow an object to be used in loops. __iter__() returns the iterator, while __next__() returns the next value until no values remain

```python
class Counter:
    def __init__ (self,limit):
        self.limit = limit
        self.current = 1
    def __iter__ (self):
        return self
    def __next__ (self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
for num in Counter(5):
    print(num)
```


# **___bool__ - Boolean Evaluation_**

The __bool__() method defines whether an object should be considered True or False

```python
class BankAccount:
    def __init__ (self,balance):
        self.balance = balance
    def __bool__ (self):
        return self.balance>0
account = BankAccount(500)
if account:
    print("Account is active")
```


# **___del__() - Destructor_**

The __del__() method is called when an object is about to be destroyed. It is mainly used for clean-up operations

```python
class Resource:
    def __init__ (self,):
        print("Resource allocated")
    def __del__ (self):
        print("Resource released")
r = Resource()
```
